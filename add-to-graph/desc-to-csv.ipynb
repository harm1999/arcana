```python
import json
with open("../replication-package/jpacman-goals_3-gpt-3.5-turbo.json") as f:
    json_data = json.load(f)
```

```python
import csv
```

```python
header = ('id', 'description')
```

```python
rows = []

for pkg in json_data:
	rows.append((pkg, json_data[pkg]['description']))
	for clss in json_data[pkg]['classes']:
		rows.append((clss, json_data[pkg]['classes'][clss]['description']))
		for method in json_data[pkg]['classes'][clss]['methods']:
			rows.append((f"{clss}.{method}", json_data[pkg]['classes'][clss]['methods'][method]['description']))

rows
```

```python
with open(f"jpacman-descriptions.csv", mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(header)
    writer.writerows(rows)
```
